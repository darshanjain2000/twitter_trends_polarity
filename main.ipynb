```python
import tkinter as tk 
from tkinter import ttk
import tweepy
import pandas as pd
import re
from textblob import TextBlob

win = tk.Tk()
win.title("Twitter Santiments")


CONSUMER_KEY='XXXXX'
CONSUMER_SECRET='XXXXXXX'
ACCESS_TOKEN='XXXXXXXX'
ACCESS_TOKEN_SECRET='XXXXXXXXXX'

# Authenticate to Twitter
auth = tweepy.OAuthHandler(CONSUMER_KEY, CONSUMER_SECRET)
auth.set_access_token(ACCESS_TOKEN, ACCESS_TOKEN_SECRET)


# To check if authentication is ok or not
def checkConnection():
    api = tweepy.API(auth)
    try:
        api.verify_credentials()
        print("Authentication OK")
    except:
        print("Error during authentication")
# checkConnection()


# def hashtags():
def clean_tweet(tweet): 
        ''' 
        Utility function to clean tweet text by removing links, special characters 
        using simple regex statements. 
        '''
        return ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)", " ", tweet).split()) 
    
def sentimentsCheck():
    # API object
    api = tweepy.API(auth, wait_on_rate_limit=True)

#     i=0
    df=pd.DataFrame(columns=['User_Name','Polarity' ,'Tweet'])
    
    # Extract tweets with hasgtag
    hashtagVal=hashtag.get()
    for tweet in api.search(q=hashtagVal, count=10):
        clean=clean_tweet(tweet.text)
        pol=TextBlob(clean)
        if pol.sentiment.polarity >=0: 
            sentiments='positive'
        else: 
            sentiments='negative'
        df=df.append({'User_Name': tweet.user.name,'Polarity':sentiments, 'Tweet': clean}, ignore_index=True)
#         i=i+1
#         print(i ,tweet.user.name,"\n", tweet.text ,"\n")
    print(df)
    

ttk.Label(win, text="Enter a #:").grid(column=0, row=0)
hashtag = tk.StringVar()
hashtagEnter = ttk.Entry(win, width=40, textvariable=hashtag)
hashtagEnter.grid(column=1, row=0)

action = ttk.Button(win, width=20,text="Check!!", command=sentimentsCheck)
action.grid(column=1, row=1)

hashtagEnter.focus()

ttk.Label(win,text="Result").grid(column=0,row=3)
# result = tk.StringVar() # 6
# resultEnter = ttk.Entry(win, width=60, textvariable=result).grid(column=0,row=4)

win.mainloop()
```
